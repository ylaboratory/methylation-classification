# mlcm by system

```python
_, meta = dill.load(open(f'./../data/GEO/preprocessed/training.dill', 'rb'))
```

```python
meta
```

```python
with open('./../figures/system_to_children.dill', 'rb') as f:
    system_to_children = dill.load(f)
```

```python
with open("./../figures/mlcm_id.dill", "rb") as f:
    true_id, pred_id = dill.load(f)
```

```python
system_to_confmat = dict()
system_to_normal_confmat = dict()

for system_node, children in system_to_children.items():
    
    children_samples = set(meta[meta['training ID'].isin(children)].index)
    children_idx = [x for x in true_id.index if x.split('.')[1] in children_samples]
    
    system_true = true_id.loc[children_idx]
    system_pred = pred_id.loc[children_idx]
    
    system_conf_mat, system_normal_conf_mat = mlcm.cm(system_true.values,system_pred.values, False)
    
    system_to_confmat[system_node] = system_conf_mat
    system_to_normal_confmat[system_node] = system_normal_conf_mat
```

```python
with open("./../figures/system_to_confmats.dill", "wb") as f:
    dill.dump((system_to_confmat, system_to_normal_confmat), f)
```
